define a state vector then verify its normalization = 1 because the valid quntum state norm = 1

```python
import math
import numpy as np

a = 1 / math.sqrt(3)
b = math.sqrt(2/3)


state_vec = [a , b]

probs = np.abs(state_vec)**2

print("the state vector of matter :",state_vec)

print("______________")

print("the state of matter original probabilty: ")
print("______________")
print('prob1 = ', probs[0])
print("---")
print('prob2 = ', probs[1])
print("______________")



norm = a**2 + b**2

print("The Normalization Verfication : ",norm)
```

Now to evolve the state with U = Unitary operator = |s1> = U |s> (s = state)

```python
# defined a theta so the u can rotate the states
theta = np.pi / 4
# activated the theta with cos and -sin in arr 0 and sin and cos in arr[1] this will be activtion for the rotation of the quantom vector state
U = np.array([[np.cos(theta), -np.sin(theta)],[np.sin(theta), np.cos(theta)]])


# |s1> = U |s> (s = state)
new_state = U @ state_vec

new_state_probs = np.abs(new_state)**2

print("(always)theta_is = ",theta)
print("_____________________")


print("old state :  ", state_vec)

print("_____________________")
print("The PROBS of (old) state of matter = ",probs)
print("_____________________")


print("new_state ",new_state)
print("_____________________")
print("The PROBS of (new) state of matter = ",new_state_probs)
print("_____________________")

# and like that we evovled The state vec with time
```

lets evolve a  state for s in range of 10

```python
state_vec = [a, b]
history = [state_vec] # Store the initial state

for s in range(10000):
  new_state = U @ state_vec
  state_vec = new_state # Update state_vec with the new_state for the next iteration
  history.append(state_vec) # Store the new state
  print(f"step {s}: {state_vec}")

history = np.array(history) # Convert list to numpy array for easier plotting
```

```python
# Define entangled state (Bell state)
state = {
    '00': 1/np.sqrt(2),
    '11': 1/np.sqrt(2)
}

# Probabilities
probs = {
    '00': np.abs(state['00'])**2,
    '11': np.abs(state['11'])**2
}

# Choose one outcome
measured_state = random.choices(list(probs.keys()), weights=list(probs.values()))[0]

# Collapse both qubits
collapsed_state = measured_state

print(f"Measured entangled state: {measured_state}")
print(f"After collapse: Qubit A = {collapsed_state[0]}, Qubit B = {collapsed_state[1]}")
```
